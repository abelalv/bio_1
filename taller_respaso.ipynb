6. Las bacterias a menudo se cultivan en placas de agar y forman colonias circulares. El área de una colonia es proporcional al número de bacterias que contiene. + Derivar la recursión para el tamaño de la población.

En este ejercicio, estamos modelando el crecimiento de una colonia de bacterias en una placa de agar. La clave es que la tasa de crecimiento de la población de bacterias depende del número de bacterias en el borde de la colonia, lo cual se relaciona con la circunferencia de la colonia.

### Paso 1: Derivar la ecuación recursiva
- La ecuación dada es $ N_{t+1} = N_t + I $, donde $ I $ es la cantidad de nuevos individuos que se añaden a la población en el tiempo $ t $.
- $ I $ es proporcional a la circunferencia de la colonia. Como la colonia es circular, su circunferencia es $ C = 2\pi r $, y el área $ A $ es proporcional al número de bacterias.
- Asumiendo que el número de bacterias es proporcional al área ($ A \propto N_t $), el radio $ r $ es proporcional a $ \sqrt{N_t} $, y por lo tanto, la circunferencia $ C $ es proporcional a $ \sqrt{N_t} $.
- Esto significa que $ I $, el número de nuevos individuos, es proporcional a $ \sqrt{N_t} $, con una constante de proporcionalidad $ R $.

Así, la ecuación recursiva para el tamaño de la población en el tiempo $ t+1 $ se convierte en:
$$ N_{t+1} = N_t + R \cdot 2 \cdot \sqrt{\pi N_t} $$

### Paso 2: Simulación del crecimiento
Para simular el crecimiento de la población, consideramos los siguientes parámetros:
- Población inicial: $ N_0 = 1000 $ bacterias
- Constante de proporcionalidad: $ R = 0.5 $
- Número de pasos de tiempo: $ 10 $ meses

La simulación se realiza mediante el siguiente código en Python:



```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
R = 0.5  # Proportionality constant
initial_population = 1000  # Initial population size
timesteps = 10  # Number of time steps

# Initialize population array
population = np.zeros(timesteps + 1)
population[0] = initial_population

# Compute population over time
for t in range(1, timesteps + 1):
    circumference = 2 * np.sqrt(np.pi * population[t-1])  # Proportional to the square root of the population
    I = R * circumference
    population[t] = population[t-1] + I

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(range(timesteps + 1), population, marker='o')
plt.title("Bacteria Population Over Time")
plt.xlabel("Time (months)")
plt.ylabel("Population Size")
plt.grid(True)
plt.show()
```
