# Límite funciones


Se dice que $\lim_{x\to a}f(x)=L$ si 
para cada $\epsilon>0$ existe un $\delta>0$ tal que si $0<|x-a|<\delta$ entonces $|f(x)-L|<\epsilon$.

# Cálculo de límites 

Si $f(x)$  es un polinomio de grado $n$ entonces $\lim_{x\to a}f(x)=f(a)$.

Si $f(x)=\frac{p(x)}{q(x)}$ y $q(a)\neq 0$ entonces $\lim_{x\to a}f(x)=f(a)$.

Si $f(x)=g(x)h(x)$ entonces $\lim_{x\to a}f(x)=\lim_{x\to a}g(x)\lim_{x\to a}h(x)$.

# Propíedades de límites 


1. Si $\lim_{x\to a}f(x)=L$ y $\lim_{x\to a}g(x)=M$ entonces $\lim_{x\to a}(f(x)+g(x))=L+M$.

2. Si $\lim_{x\to a}f(x)=L$ y $\lim_{x\to a}g(x)=M$ entonces $\lim_{x\to a}(f(x)-g(x))=L-M$.

3. Si $\lim_{x\to a}f(x)=L$ y $\lim_{x\to a}g(x)=M$ entonces $\lim_{x\to a}(f(x)g(x))=LM$.



# Definición de una función continua 

Una función $f(x)$ es continua en $a$ si $\lim_{x\to a}f(x)=f(a)$.

# Propiedades de funciones continuas

1. Si $f(x)$ y $g(x)$ son continuas en $a$ entonces $f(x)+g(x)$ y $f(x)g(x)$ son continuas en $a$.

2.  Si $f(x)$ y $g(x)$ son continuas en $a$ y $g(a)\neq 0$ entonces $\frac{f(x)}{g(x)}$ es continua en $a$.


# Definición de una función continua en un intervalo

Una función $f(x)$ es continua en un intervalo $[a,b]$ si es continua en cada punto del intervalo.



# limite de una función continua

Si $\lim_{x\to a }g(x)=L$ y $f$ es continua en $L$ entonces $\lim_{x\to a}f(g(x))=f(L)$.


# Teorema de una continuidad de una función compuesta

Si $f(x)$ es continua en $a$ y $\lim_{x\to a}g(x)=L$ entonces $\lim_{x\to a}f(g(x))=f(L)$.

# Teorema del sandwich

Si $f(x)\leq g(x)\leq h(x)$ para todo $x$ en una vecindad de $a$ excepto posiblemente en $a$ y si $\lim_{x\to a}f(x)=\lim_{x\to a}h(x)=L$ entonces $\lim_{x\to a}g(x)=L$.

# Ejemplo 

Calcular $\lim_{x\to 0}x^2\sin\left(\frac{1}{x}\right)$.




```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x values
x = np.linspace(-1, 1, 10000)

# Calculate the corresponding y values
y = x**2*np.sin(1/x)
y_1= x**2


# Plot the function
plt.plot(x, y)
plt.plot(x, y_1)
plt.plot(x, -y_1)
plt.xlabel('x')
plt.ylabel('sin(1/x)')
#plt.xlim(-0.1, 0.1)
plt.title('Plot of sin(1/x)')
plt.show()
```
